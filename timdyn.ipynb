```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function t1(u, e, d, gamma, lambda)
def t1(summ, gamma, lambd):
    return 1 / gamma * np.log((summ) / lambd)

# Define constants for u, e, d, gamma, lambda
gamma = 1
lambda_val = 10

# Generate sample values for u + e + d
sum_u_e_d_vals = np.arange(10, 10001, 1)  # Adjust the range based on your needs

# Calculate t1 values for different sum_u_e_d values
t1_values = t1(sum_u_e_d_vals, gamma, lambda_val)
# Remove negative t1 values
t1_values = np.maximum(t1_values, 0)  # Setting negative values to 0

# Plot t1 as a function of u + e + d (showing only positive values)
plt.figure(figsize=(8, 6))
plt.plot(sum_u_e_d_vals, t1_values, label='$t_1(u, e, d)$', color='blue')
plt.xlabel('u + e + d')
plt.ylabel('$t_1$')
#plt.title('Behavior of $t_1$ as a function of u + e + d for $gamma=1~and~\lambda=10$')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function t2(u, e, d, alpha, beta, gamma, lambd)
def t2(u, e, d, alpha, beta, gamma, lambd):
    denominator = beta * (alpha * u - alpha * gamma / beta + d)
    numerator = np.log(u / (gamma / beta)) * (alpha * ((u - gamma / beta) / d) + 1)
    return numerator / denominator

# Define constants for alpha, beta, gamma, lambda
alpha = 0.4
beta = 3
gamma = 3
lambda_val = 10

# Generate sample values for u and d
u_vals = np.arange(gamma / beta, 5001, 10)

# Calculate t2 values for different u values
t2_values_u = t2(u_vals, None, lambda_val, alpha, beta, gamma, lambda_val)

# Plot t2 as a function of u
plt.figure(figsize=(8, 6))
plt.plot(u_vals, t2_values_u, label='$t_2(u, e, d)$', color='blue')
plt.xlabel('u')
plt.ylabel('$t_2$')
#plt.title(r'$t_2$ as a function of u for $\gamma=\beta$ and $d=10$')
plt.legend()
plt.grid(True)
plt.show()
```
